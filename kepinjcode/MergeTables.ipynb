```python
#create new pandaframe to store the merged table
new_columns = collect_blsanal.columns.values
merged = pd.DataFrame(np.zeros_like(overlap_inj), columns=collect_blsanal.columns)
merged["Origin"] =""
```

```python
#period error function
def period_func(vals, p_bls, p_inj):
    a, b = vals
    a = np.rint(a)
    b = np.rint(b)   
    return np.abs((a*p_bls - b*p_inj) / 0.5/(a*p_bls + b*p_inj))
```

```python
from scipy.optimize import minimize, fmin, fminbound, fmin_slsqp, brute

start = time.time()
for i, label in enumerate(overlap_inj["tag"]):
    #print label
    #print overlap_inj[["BLS_SignaltoPinknoise", "MES_scaled", "MES_expect","period_inj", "BLS_Period"]].values[np.where(overlap_inj["tag"]==label )[0]]
    #print collect_blsanal_koi[["BLS_SignaltoPinknoise", "MES_scaled", "period_inj", "BLS_Period"]].values[np.where(collect_blsanal_koi["tag"]==label )[0]]
    #for the koi files, i'd imagine the inj period is the koi period?
    
    #how many signals per koi 
    signal_number = collect_blsanal_koi.values[np.where(collect_blsanal_koi["tag"]==label )[0]].shape[0]
    #print signal_number
    #inner loop to account for KOIS with multiple signals (multiple planets)
    for j in range(signal_number):
        val1 = collect_blsanal_koi["period_inj"].values[np.where(collect_blsanal_koi["tag"]==label )[0]][j]
        val2 = collect_blsanal_koi["BLS_Period"].values[np.where(collect_blsanal_koi["tag"]==label )[0]][j]
        snr = collect_blsanal_koi["BLS_SignaltoPinknoise"].values[np.where(collect_blsanal_koi["tag"]==label )[0]][j]
        rranges = (slice(1,6, 1), slice(1,6, 1))
        results = brute(period_func, ranges=rranges, args=(val1, val2 ), full_output=True, finish= None)
        tol = period_func(results[0],val1, val2)
        #print tol, results[0]
        if tol<0.02 and (merged["BLS_SignaltoPinknoise"][i]<= snr ):
            merged.iloc[i, :-1] = collect_blsanal_koi.values[np.where(collect_blsanal_koi["tag"]==label )[0]][j].astype(float)
            merged.iloc[i, -1] = "koi"

    
    #injected period detection
    val1 = overlap_inj["period_inj"].values[np.where(overlap_inj["tag"]==label )[0]]
    val2 = overlap_inj["BLS_Period"].values[np.where(overlap_inj["tag"]==label )[0]]
    snr = overlap_inj["BLS_SignaltoPinknoise"].values[np.where(overlap_inj["tag"]==label )[0]][0]

    results = brute(period_func, ranges=rranges, args=(val1, val2 ), full_output=True, finish= None)
    tol = period_func(results[0],val1, val2)
    #print tol, results[0]
    #if the koi isnt detected, then take the injected signal as long as its higher
    #this will take care of the case if the koi has a large period distance above the threshold
    if (snr >= merged["BLS_SignaltoPinknoise"][i] ) :
        merged.iloc[i, :-1] = overlap_inj.values[np.where(overlap_inj["tag"]==label )[0]][0].astype(float)
        merged.iloc[i, -1] = "inj"
    
    
    #merged.head()
    #break
end= time.time()
print "Time taken %f" % (end-start)
print len(np.where(merged["Origin"]=="inj")[0])
print len(np.where(merged["Origin"]=="koi")[0])
merged.head(10)
```
