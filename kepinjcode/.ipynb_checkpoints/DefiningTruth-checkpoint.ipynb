```python
import numpy as np
import pandas as pd
import time
import matplotlib
from matplotlib import pyplot as plt
```

The merged.csv is only 1514, as its the overlap of the KOIs and their respective injected files. Here I then merge it with the rest of the injected files that did not over lap with the KOIs. To make that clear, in the Origin column where I keep traack of where the files come from, I give it a "INJ" label. 

```python
inj = pd.read_csv("collectblsanal.csv", index_col=False)
merged = pd.read_csv("merged.csv")
inj["tag"] = (inj["kepid"].astype(str) + inj["segno"].astype(str)).astype(int)
```

```python
merged["kepid"] = merged["kepid"].astype(int)
```

```python
inj["Origin"] = "INJ"
inj.head()
```

```python
print len(np.where(np.in1d(inj["tag"].values, merged["tag"].values) ==True)[0])
print merged.shape
print inj.shape
```

```python
just_inj= inj[np.in1d(inj["tag"].values, merged["tag"].values) ==False]
final_table = pd.concat([just_inj, merged] )
print final_table.shape
#final_table.head()
```

```python
#primary1 feat files
inj_feat_files = pd.read_csv("../../TransitClassificationRobovetter/nhussain_scripts/Data/Kplr_feat_files_34000.csv", index_col=False)
inj_feat_files = inj_feat_files.drop("Unnamed: 0", axis  =1)
inj_feat_files[["star_id", "segment"]] = inj_feat_files[["star_id", "segment"]].astype(int)
inj_feat_files["tag"] = (inj_feat_files["star_id"].astype(str) + inj_feat_files["segment"].astype(str)).astype(int)
inj_feat_files["detect_distance"] =0 #this will be fixed afterwards. #this is to make it easier to concatenate to the final feature table
inj_feat_files.head()
```

```python
#take the feats from the feat files and keep in the same order as the final table of the koi /inj signals
start = time.time()
final_feats = pd.DataFrame(np.zeros_like(inj_feat_files), columns=inj_feat_files.columns.values)
for i, label in enumerate(final_table["tag"]):
    #print label
    #print inj_feat_files[inj_feat_files["tag"] == label].values
    try:
        final_feats.iloc[i, :] = inj_feat_files[inj_feat_files["tag"] == label].values
        final_feats.iloc[i, -1] = final_table["detect_distance"].values[i]
        #print final_table["detect_distance"].values[i]
        #break
    except Exception as e:
        print i, label
        print e
        break
    #break
end = time.time()
print "Time taken: %f" % (end - start)
final_feats.head(10)
```

```python
#df.loc[~(df==0).all(axis=1)]
final_feats= final_feats.loc[~(final_feats==0).all(axis = 1)]
print final_feats.shape
```

Need to now define a truth column, will do what I have been doing before
signal cut, transit cut, and period detected cut
Note: get injected period in the collected inj files - this alreayd has done the period cut for us.

```python
period_cut = np.asarray(final_feats["detect_distance"]!=-1)
print "Period Cut:", len(np.where(period_cut ==True)[0]), len(np.where(period_cut ==False)[0]) 

transit_cut = final_table["period_inj"]<26./3.
print "Transit Cut:", len(np.where(transit_cut ==True)[0]), len(np.where(transit_cut ==False)[0]) 

noise_cut = final_table["MES_scaled"]>6 # this is low
print "Noise Cut", len(np.where(noise_cut ==True)[0]), len(np.where(noise_cut ==False)[0]) 

total_cut = period_cut*transit_cut*noise_cut
print "Total Cut", len(np.where(total_cut ==True)[0]), len(np.where(total_cut ==False)[0]) 

#np.savetxt("truth_values.txt", total_cut)
```
